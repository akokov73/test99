```python
import regex
def clear_text (text_hist):
    text_hist = text_hist.replace(':',': ')    
    text_hist = regex.subf("(([А-ЯЁ][а-яё])\w+)", " {1}", text_hist)  
    text_hist = regex.subf(r"(([А-ЯЁ][а-яё]?)\w+)", " {1}", text_hist)    
    text_hist = regex.subf(r"(([А-ЯЁ][а-яё])\w+)(([А-ЯЁ][а-яё])\w+)", " {1} {3}", text_hist)       
    text_hist = regex.subf(r"(([А-ЯЁ])\w+)(([А-ЯЁ][а-яё])\w+)", " {1} {3}", text_hist)           
    text_hist = regex.subf("(([А-ЯЁ][а-яё]))", " {1}", text_hist)  
    text_hist = text_hist.replace(' .','.')
    text_hist = text_hist.replace('..','.')
    text_hist = text_hist.replace('  ',' ')
    text_hist = text_hist.replace('  ',' ')
    text_hist = text_hist.replace('  ',' ')
    text_hist = text_hist.replace('  ',' ') 
    text_hist = text_hist.replace(',. ',', ')
    text_hist = text_hist.replace('..','.') 
    text_hist = text_hist.replace('..','.')
    
    text_hist = text_hist.replace('.',' ')     
    text_hist = text_hist.replace(',',' ')     
    text_hist = text_hist.replace('=',' ')     
    text_hist = text_hist.replace('/',' ')     
    text_hist = text_hist.replace('0',' ')     
    text_hist = text_hist.replace('1',' ')     
    text_hist = text_hist.replace('2',' ')     
    text_hist = text_hist.replace('3',' ')     
    text_hist = text_hist.replace('4',' ')     
    text_hist = text_hist.replace('5',' ')     
    text_hist = text_hist.replace('6',' ')     
    text_hist = text_hist.replace('7',' ')     
    text_hist = text_hist.replace('8',' ')     
    text_hist = text_hist.replace('9',' ')     
    text_hist = text_hist.replace('5',' ')     
    text_hist = text_hist.replace('(',' ')     
    text_hist = text_hist.replace(')',' ')     
    text_hist = text_hist.replace(':',' ')     
    return text_hist
```
